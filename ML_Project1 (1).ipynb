```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy="median")
demond_clear=data.drop(columns =['clarity','color','cut'],axis=1)
imputer.fit(demond_clear)
imputer.statistics_
X = imputer.transform(demond_clear)
demond_train = pd.DataFrame(X, columns=demond_clear.columns,
index=demond_clear.index)
demond_train
```

```python
#Dropping the outliers. 
data = demond_train[(demond_train["depth"]<75)&(demond_train["depth"]>45)]
data = demond_train[(demond_train["table"]<80)&(demond_train["table"]>40)]
data = demond_train[(demond_train["x"]<30)]
data = demond_train[(demond_train["y"]<30)]
data = demond_train[(demond_train["z"]<30)&(demond_train["z"]>2)]
data.shape
```
